## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m2_survey_data.csv")
df.dropna(subset=['CompFreq', 'CompTotal', 'ConvertedComp'], inplace=True)
def fillcomp(freq,total):
    if freq=='Yearly':
        return total
    elif freq=='Monthly':
        return total * 12
    elif freq=='Weekly':
        return total * 50
df['ConvertedComp']=df.apply(lambda row:fillcomp(row['CompFreq'], row['ConvertedComp']),axis=1)
# your code goes here
import numpy as np
import pandas as pd
import scipy.stats as stats
%matplotlib inline
df['ConvertedComp'].plot(kind='hist')
plt.title=('Histogram of ConvertedComp')
plt.xlabel('ConvertedComp')
plt.ylabel('Count')
plt.show()

```

### Box Plots


Plot a box plot of `Age.`

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd
import matplotlib.pyplot as plt  # Import the matplotlib library

# Your SQL query
QUERY = """
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""
df = pd.read_sql_query(QUERY, conn)
df.set_index('DatabaseDesireNextYear', inplace=True)
color_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen']
labels = df.index
# Create the pie chart
plt.figure(figsize=(10, 6))  # Set the figure size
plt.pie(df['Count'], labels=labels, colors=color_list, autopct='%1.1f%%', startangle=90, shadow=True, pctdistance=0.85)
plt.title('Top 5 Database Desire Next Year')  # Add the title
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Add legend with proper labels
plt.legend(labels, loc='upper left')

# Show the pie chart
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age IN (30,31,32,33,34,35)
"""
df = pd.read_sql_query(QUERY, conn)

df=df.groupby(by='Age')['WorkWeekHrs','CodeRevHrs'].median()

df.plot(kind='area')
plt.title('Stacked Chart of Median WorkWeekHrs & CodeRevHrs for ages 30-35')
plt.ylabel('Age?')
plt.xlabel('Median Hrs')
plt.show()


```

Plot the median `ConvertedComp` for all ages from 45 to 60.

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python


QUERY = """
SELECT MainBranch,COUNT(MainBranch) as count_of_MainBranch
FROM master
group by MainBranch
order by count_of_MainBranch
"""

#the read_sql_query runs the sql query and returns the data as a dataframe
df9 = pd.read_sql_query(QUERY,conn)

# plot horizontal bar chart using column MainBranch
df9.plot.barh(figsize=(6, 3)) 

# add xlabel and title

plt.xlabel('count_of_MainBranch') 
#labels=["I am a developer by profession","I am not primarily a developer, but I write code sometimes as part of my work"]
#plt.yticks([1, 0], labels)
plt.title('horizontal bar chart using column MainBranch')
  

# Show the plot
plt.show()
```
